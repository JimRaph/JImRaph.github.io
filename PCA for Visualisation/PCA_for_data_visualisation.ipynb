WORKING WITH PCA FOR DATA VISUALIZATION


```python
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
```

```python
data = load_breast_cancer()
```

```python
dir(data)
```

```python
data.data.shape
```

WE HAVE 30 FEATURES AND 569 SAMPLES TO WORK WITH


```python
#shorting what I have to type
data_val = data.data

```

```python
#view some of the data
data_val[:5]
```

```python
#get feature names for each feature
data_val_name = data.feature_names
data_val_name
```

```python
data_target = data.target
data_target.shape
```

```python
#add target column to the feature columns
#first reshape target shape
data_target = np.reshape(data_target, (569,1))
#concatenate the target to the others
data_full = np.concatenate([data_val, data_target], axis=1)
#confirm changes
data_full.shape
```

```python
#we have feature names but no target name, changing this
data_target_name = 'target_label'
data_target_name
```

```python
dataset_names = np.append(data_val_name,data_target_name)
dataset_names
```

```python
#convert dataset into pandas dataframe for easier handling
dataset = pd.DataFrame(data_full, columns = dataset_names)
dataset.head()
```

```python
#check ratio of values for the target_label
print(dataset.target_label.value_counts())
```

```python
#PCA is greatly influenced by data scale, to fight against this, scaling is applied
from sklearn.preprocessing import StandardScaler

#initializing the scaler
scaler = StandardScaler()

#apply scaler to data

data_scaled = scaler.fit_transform(dataset.loc[:, data_val_name])
```

```python
type(data_scaled)
```

```python
#I should have scaled before attaching the target_label, I will just add again, sighs
data_scaled_full = np.concatenate([data_scaled, data_target], axis=1)

#convert to dataframe again. sighs.
dataset_scaled = pd.DataFrame(data_scaled_full, columns = dataset_names)
```

```python
dataset_scaled.head()
```

HURRAY!!!


```python
#concerning the Target label column, 0 stands for benign and 1 for malignant
#making changes

dataset_scaled['target_label'].replace(0, 'Benign',inplace=True)
dataset_scaled['target_label'].replace(1, 'Malignant',inplace=True)

#confirm changes
dataset_scaled.head(2)
```

Applying PCA. I don't know the number of components that will re

```python
#initialize PCA
pca = PCA(n_components = 3)

#apply transformation
pca_data = pca.fit_transform(dataset_scaled.loc[:, data_val_name])
```

I had realised a little way back, that I will add the target label to the rest of the dataset at the point of visualisation after applying the PCA. I should delete some of the steps above to at least keep the notebook short but I'm a little lazy to do that.

Instead, we will follow the great saying; fear not the one that perform a 100 step but he that performs 3 of the same step.....or however the saying goes. point is, we will convert to dataframe and attach the label again.

sighs


```python
#again, convert to dataframe and attach target label
pca_data_full = np.concatenate([pca_data, data_target], axis=1)

pca_data_df = pd.DataFrame(pca_data_full, columns = ['comp_1','comp_2','comp_3','target_label'])

#pca_data_df['target_label'].replace(0, 'Benign',inplace=True)
#pca_data_df['target_label'].replace(1, 'Malignant',inplace=True)

pca_data_df.head()
```

```python
import matplotlib.pyplot as plt
```

3D visualization

```python
fig = plt.figure()

ax = fig.add_subplot( projection='3d')

ax.scatter(pca_data_df['comp_1'], pca_data_df['comp_2'], pca_data_df['comp_3'],marker='o',c=pca_data_df['target_label'])

plt.show()
```

```python
#check ratio of variance for the components

print('The explained variance ratio for each of the components(comp_1,comp_2,comp_3): {}'.format(pca.explained_variance_ratio_))
```

It appears that, the third component contains only 9% of the variance with the first taking the lion share.

2D visualisation

```python
#initialize PCA for 2 components
pca_2 = PCA(n_components = 2)

pca_data_2 = pca_2.fit_transform(dataset_scaled.loc[:, data_val_name])

pca_data_full_2 = np.concatenate([pca_data_2, data_target], axis=1)

pca_data_df_2 = pd.DataFrame(pca_data_full_2, columns = ['comp_1','comp_2','target_label'])
```

```python
pca_data_df_2.head(3)
```

```python
#plt.figure(figsize=(7,7))

indices0 = pca_data_df['target_label'] == 0
indices1 = pca_data_df['target_label'] == 1
plt.scatter(pca_data_df_2.loc[indices0, 'comp_1'], pca_data_df_2.loc[indices0, 'comp_2'], c = 'red')
plt.scatter(pca_data_df_2.loc[indices1, 'comp_1'], pca_data_df_2.loc[indices1, 'comp_2'], c = 'g')
plt.legend(['Benign','Malignant'])
plt.show()
```

```python
print('Again, checking explained variance ratio of the two components')
print('The Explained Variance Ratio: {}'.format(pca_2.explained_variance_ratio_))
```
